```python
import geopandas as gpd

# Đường dẫn tới file shapefile chứa dữ liệu dân số
population_data_path = "CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp"

# Đọc dữ liệu shapefile
population_gdf = gpd.read_file(population_data_path)
population_gdf
```

## Step4: User clicks

```python
user_clicks_path = "/content/click_data/click_data.json"

user_clicks_gdf = gpd.read_file(user_clicks_path)
user_clicks_gdf
```

## Step5: Lọc ra 10 quận hot nhất

```python
# Tính tỷ lệ tăng dân số
population_gdf['growth_rate'] = population_gdf['Pop_2019'] / population_gdf['Pop_2017']

# Lọc ra 10 quận huyện có tốc độ tăng nhanh nhất
top_10_districts = population_gdf.sort_values(by='growth_rate', ascending=False).head(10)
print(top_10_districts[['Dist_Name', 'growth_rate']])
```

## Step6: Join 2 bảng -> filter_clicks table

```python
from geopandas.tools import sjoin

# Ensure both GeoDataFrames are in the same CRS
top_5_districts = top_10_districts
print("-> District CRS:", top_5_districts.crs)
print("-> User Clicks CRS:", user_clicks_gdf.crs)

# If CRS are different, reproject user clicks to match district boundaries
if top_5_districts.crs != user_clicks_gdf.crs:
    user_clicks_gdf = user_clicks_gdf.to_crs(top_5_districts.crs)
print("-> District CRS:", top_5_districts.crs)
print("-> User Clicks CRS:", user_clicks_gdf.crs)

# Perform a spatial join to find clicks within the top 5 districts
filtered_clicks = sjoin(user_clicks_gdf, top_5_districts, how='inner', predicate='within')

# Check the result
print(f"Number of matching points: {len(filtered_clicks)}/{len(user_clicks_gdf)}")
filtered_clicks
```

## Step7: Kmean(20, 42) để phân cụm cho các row của filter_clicks table

```python
from sklearn.cluster import KMeans

# Lấy tọa độ từ dữ liệu click
coordinates = filtered_clicks.geometry.apply(lambda point: (point.x, point.y)).tolist()

# Chạy KMeans với K=20
kmeans = KMeans(n_clusters=20, random_state=0)
filtered_clicks['cluster'] = kmeans.fit_predict(coordinates)

filtered_clicks
```

## Step8: SQL - Tìm cụm chiếm đa số cho mỗi quận -> largest_clusters table

```python
filtered_clicks.groupby(['Dist_Name', 'cluster'])
tmp = filtered_clicks.groupby(['Dist_Name', 'cluster']).size().reset_index(name='count')
# tmp = tmp.groupby('Dist_Name').max('count')
print("assert:", sum(tmp['count']))
tmp
```

```python
if 'cluster' in filtered_clicks.columns:
    largest_clusters = (
        filtered_clicks.groupby(['Dist_Name', 'cluster'])
        .size()
        .reset_index(name='count')
        .sort_values(['Dist_Name', 'count'], ascending=[True, False])
        .groupby('Dist_Name')
        .first()
        .reset_index()
    )
    print(largest_clusters)
    print("Number clicks of big-clusters:", sum(largest_clusters['count']))
else:
    print("Error: 'cluster' column is missing.")
```

## Step9-10: Plot largest_clusters recorve ...

```python
largest_cluster_points = filtered_clicks.merge(
    largest_clusters[['Dist_Name', 'cluster']],
    on=['Dist_Name', 'cluster']
)
print("assert:", len(largest_cluster_points))
largest_cluster_points
```

```python
import matplotlib.pyplot as plt

# Extract x (longitude) and y (latitude) coordinates
x_coords = largest_cluster_points.geometry.x
y_coords = largest_cluster_points.geometry.y

# Plot the heatmap using Matplotlib
plt.figure(figsize=(30, 30))
population_gdf.plot(color='none', edgecolor='black')

plt.hexbin(x_coords, y_coords, gridsize=50, cmap='Reds', mincnt=1)

# Add a color bar
plt.colorbar(label='Number of Points')

# Set labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Heatmap of Largest Clusters per District')

# Save the heatmap as a PNG file
plt.savefig('heatmap.png', format='png', dpi=300)

# Show the plot (optional)
plt.show()
```
